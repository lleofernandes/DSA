### Representação Gráfico de Funções
Vamos criar um exemplo com a função quadrática 
$$ 
f(x) = x^2
$$ 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return x ** 2 
```

```python
# cria um conjunto de valores aleatórios de x (por exemplo, de -10 a 10)
x = np.linspace(-10, 10, 400)

# calcula os valores de y para cada valor de x
y = f(x) 
```

```python
# Cria o gráfico
plt.figure(figsize=(6, 4))
plt.plot(x, y, label='f(x) = x^2', color='blue')
plt.title('Gráfico da Função Quadrática')
plt.xlabel('x')
plt.ylabel('f(x)') 

plt.legend()

plt.grid(True)
plt.show()
```

### Seno, Cosseno e Tangente
- O seno é uma função matemática fundamental comumente encontrada na trigonometria, um ramo da matemática que lida com as relações entre os lados e os ângulos dos triângulos. 
- O Seno é a razão entre o cateto oposto e a hipotenusa de um triângulo retângulo, ou seja, trata-se de uma razão trigonométrica

```python
import math

# Calculando o seno de 30 graus

angulo_graus = 30 # Define o angulo em graus
angulo_radianos = math.radians(angulo_graus) # Converte o angulo de graus para radianos
seno = math.sin(angulo_radianos) # Calcula o seno do angulo, que agora está em radianos
print(f"O seno de {angulo_graus} graus é: {seno:.2f}") # exibe o valor do seno
```

O Cosseno de um ângulo em triangulo retângulo é a razão entre o comprimento do lado Adjacente ao ângulo e comprimento da Hipotenusa

```python
# calculando o cosseno de 45 graus
angulo_graus = 45
angulo_radianos = math.radians(angulo_graus)
cosseno = math.cos(angulo_radianos) 
print(f"O cosseno de {angulo_graus} graus é: {cosseno:.2f}")
```

A Tangente de um ângulo em um triângulo retângulo é a razão entre o comprimento do lado oposto ao ângulo e comprimento do lado Adjacente

```python
# Calculando a tangente de 60 graus
angulo_graus = 60
angulo_radianos = math.radians(angulo_graus)
tangente = math.tan(angulo_radianos)
print(f"A tangente de {angulo_graus} graus é: {tangente:.2f}")
```

Nos modelos "Transformers", utilizados em Processamento de Linguagem Natural e outras tarefas de aprendizado de maquinas como LLMs (Large Language Models)

O artigo original que introduziu os Transformers (["Attention Is All You Need"](https://arxiv.org/pdf/1706.03762.pdf) por Vaswani et al., 2017) propõe o uso de funções Seno e Cosseno com frequências variadas para esses codificadores de posição.

Por que Seno e Cosseno?
- Uniformidade e Diferenciação: AS funções Seno e Cosseno variam de maneira regular e previsível, o que ajuda o modelo a aprender facilmente a relação entre as posições
- Generalização para Sequências Longas: Essas funções permitem que o modelo generalize para posições que não foram vistas durante o treinamento, pois os padrões de Seno e Cosseno  são consistentes e se repetem.
- Preservação da Relação linear: A combinação de Seno e Cosseno, permite preservar a relação linear com a distância entre as palavras, o que é importante para o mecanismo de atenção do modelo.



### Logaritmos e Logaritmos Naturais

```python
import numpy as np

#Exemplo decalculo do logaritmo natural
x = 10
log_x = np.log(x)  # Calcula o logaritmo natural de x
print(f"O logaritmo natural de {x} é: {log_x:.2f}")


# Exemplo de cálculo do logaritmo na base 10
y = 100
log10_y = np.log10(y)  # Calcula o logaritmo na base 10
print(f"O logaritmo na base 10 de {y} é: {log10_y:.2f}")
```

```python
# Definindo numeros par ao exemmplo
a = 5
b = 3
c = 2

# Calculando logartimos

# Logaritmo do produto
log_product = math.log(a * b)

# Logaritmo do quociente
log_quotient = math.log(a / b)

# Logaritmo da potência
log_power = math.log(a ** c)

# Demonstrando as propriedades dos logaritmos
# Propriedade do produto: log(a * b) = log(a) + log(b)
print("Logaritmo do produto:", log_product, "Propriedade:", math.log(a) + math.log(b))

# Propriedade do quociente: log(a / b) = log(a) - log(b)
print("Logaritmo do quociente:", log_quotient, "Propriedade:", math.log(a) - math.log(b))

# Propriedade da potência: log(a ** c) = c * log(a)
print("Logaritmo da potência:", log_power, "Propriedade:", c * math.log(a))
```

### Exemplo de uso da transformação de Log em Data Science
Exemplo que ilustra como aplicar a transformação de Log em um conjunto de dados.
- Suponha que temos um conjunto de dados com valores que variam signitivamente e queremos aplicar uma transformação de log para normalizá-los

```python
# gerando dados de exemplo: Preco de casas
np.random.seed(0)  # Para reprodutibilidadem, ou seja, para que os resultados sejam os mesmos a cada execução
preco_casas = np.random.lognormal(mean=13, sigma = 0.6, size=1000)  # Preços em uma escala log-normal (Média = 13, Desvio Padrão = 0.6 e 1000 amostras)

# Aplicando a transformação logarítmica
preco_casas_log = np.log(preco_casas)

# Plotando os dados originais e transformados
plt.figure(figsize=(8, 4))

# Gráfico dos preços originais
plt.subplot(1, 2, 1) # uma linha, duas colunas, primeiro gráfico
plt.hist(preco_casas, bins=30, color='blue', alpha=0.7)
plt.title('Distribuição dos Preços Originais das Casas')
plt.xlabel('Preço')
plt.ylabel('Frequência')

# Gráfico dos preços transformados
plt.subplot(1, 2, 2) # uma linha, duas colunas, segundo gráfico
plt.hist(preco_casas_log, bins=30, color='green', alpha=0.7)
plt.title('Distribuição dos Preços Transformados (Logarítmica)')
plt.xlabel('Log(Preço)')
plt.ylabel('Frequência')

plt.tight_layout()
plt.show()  
```

### Exercício 2: Gráfico de uma função quadrática

Plotar um gráfico da função y = x^2 - 4x + 4

Criar uma massa de dados usando a função np.linspace()

```python
# função quadrática
x = np.linspace(-2, 6, 100)
y = x**2 - 4*x + 4  # função y = x^2 - 4x + 4

# Cria o gráfico
plt.plot(x, y, label='y = x^2 - 4x + 4', color='orange')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de uma Função Quadrática')
plt.axhline(0, color='black', lw=0.5, ls='--')  # Linha horizontal no eixo x
plt.axvline(0, color='black', lw=0.5, ls='--')  # Linha vertical no eixo y
plt.legend()
plt.grid(True)
plt.show()
```

### Exercício 6: Resolver uma equação quadrática

Escreva um script que resola a equação x^2 - 5x + 6 = 0 e imprime as raizes

```python
import cmath
# função raiz quadrada complexa x^2 - 5x + 6 = 0

# Coeficientes da equação quadrática
a, b, c = 1, -5, 6
delta = b**2 - 4*a*c

x1 = (-b + cmath.sqrt(delta)) / (2 * a)
x2 = (-b - cmath.sqrt(delta)) / (2 * a)

print(f"As raízes da equação {a}x² + {b}x + {c} = 0 são: \n x1: {x1} \n x2: {x2}")



```

### Exercício 8: Equação quadrática com discriminante


```python
a = float(input("Digite o valor de a: "))
b = float(input("Digite o valor de b: "))
c = float(input("Digite o valor de c: "))

delta = b**2 -4*a*c

x1 = (-b + cmath.sqrt(delta)) / (2 * a)
x2 = (-b - cmath.sqrt(delta)) / (2 * a)

if delta > 0:
    print(f"As raízes da equação {a}x² + {b}x + {c} = 0 são reais e distintas: \n x1: {x1} \n x2: {x2}")
    
elif delta == 0:
    print(f"As raízes da equação {a}x² + {b}x + {c} = 0 são reais e iguais: \n x1 = x2: {x1}")

else:
    print("Nenhuma raiz real")
```
